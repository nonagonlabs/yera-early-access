```python
import yera as yr
```

```python
class Widgets(yr.Struct):
    class Widget(yr.Struct):
        mass: float
        volume: float
        name: str

        def density(self):
            return self.mass / self.volume

    widgets: list[Widget]

    def total_mass(self):
        return sum(w.mass for w in self.widgets)

    def mean_density(self):
        return sum(w.density() for w in self.widgets) / len(self.widgets)

    def summary_table(self):
        import pandas as pd

        return pd.DataFrame(w.model_dump() for w in self.widgets)

    def markdown_summary(self):
        with yr.markdown("\n# Widgets:\n") as m:
            m.append("## Aggregate stats\n")
            m.append(f"  densities: {[w.density() for w in self.widgets]}\n")
            m.append(f"  total mass: {self.total_mass()}\n")
            m.append(f"  mean density: {self.mean_density()}\n")

        yr.table(self.summary_table())
```

```python
@yr.agent(name="Widget Builder")
def widget_builder() -> None:
    while True:
        prompt = yr.text_input()

        if prompt == "/quit":
            yr.quit()
            break

        if prompt.startswith("/create "):
            res = Widgets.fill(prompt[8:])
            res.markdown_summary()
        else:
            res = yr.chat(prompt)
            yr.markdown(f"response stats: {len(res)} chars long")
```

```python
widget_builder()
```
